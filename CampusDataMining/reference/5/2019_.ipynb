```python
def chose_common(borrow_book_2016,borrow_book_2017):
    user_2016=set(borrow_book_2016['user_id'].values.tolist())
    print('16年的用户数目为：{}'.format(len(user_2016)))
    user_2017=set(borrow_book_2017['user_id'].values.tolist())
    print('17年的用户数目为：{}'.format(len(user_2017)))
    book_2016=set(borrow_book_2016['book_id'].values.tolist())
    print('16年的图书数目为：{}'.format(len(book_2016)))
    book_2017=set(borrow_book_2017['book_id'].values.tolist())
    print('17年的图书数目为：{}'.format(len(book_2017)))
    common_user=list(user_2016.intersection(user_2017))
    print('16和17年的共同的用户数目为：{}'.format(len(common_user)))
    common_book=list(book_2016.intersection(book_2017))
    print('16和17年的共同的图书数目为：{}'.format(len(common_book)))
    common_user_book_2016=borrow_book_2016[borrow_book_2016['user_id'].isin(common_user)]
    common_user_book_2016=common_user_book_2016[common_user_book_2016['book_id'].isin(common_book)]
    user_2016=set(list(common_user_book_2016['user_id']))
    print('筛选之后2016年拥有的用户数目为:{}'.format(len(user_2016)))
    book_2016=set(list(common_user_book_2016['book_id']))
    print('筛选之后2016年拥有的图书数目为:{}'.format(len(book_2016)))
    common_user_book_2017=borrow_book_2017[borrow_book_2017['user_id'].isin(common_user)]
    common_user_book_2017=common_user_book_2017[common_user_book_2017['book_id'].isin(common_book)]
    user_2017=set(list(common_user_book_2017['user_id']))
    print('筛选之后2017年拥有的用户数目为:{}'.format(len(user_2017)))
    book_2017=set(list(common_user_book_2017['book_id']))
    print('筛选之后2017年拥有的图书数目为:{}'.format(len(book_2017)))
    return common_user_book_2016, common_user_book_2017
```

```python
common_user_book_19_1, common_user_book_19_2=chose_common(borrow_book_19_1, borrow_book_19_2)
```

```python
i=0
while(i<15):
    common_user_book_19_1, common_user_book_19_2=chose_common(common_user_book_19_1, common_user_book_19_2)
    i+=1
```
