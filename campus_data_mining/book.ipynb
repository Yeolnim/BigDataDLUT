```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt

df_2 = pd.read_csv("tsjyygh.txt",sep = ";", encoding = 'utf8')

df_2=df_2['sm'].value_counts()
df_2.to_csv('bookname.csv',encoding='utf-8')

df = pd.read_csv("bookname.csv",names=['sm','count'], encoding = 'utf8')
df= df.groupby(by='sm').sum()
df=df.sort_values(by='count',ascending=True)

df.to_csv('bookname.csv',encoding='utf-8')

df = pd.read_csv("bookname.csv", encoding = 'utf8')
# print(df)
df=df[-20:]


matplotlib.rcParams["font.sans-serif"] = ["simhei"]  # 黑体
matplotlib.rcParams["font.family"] = "sans-serif"
plt.barh(df['sm'],df['count'],color="green")
plt.title("Pop Books")
plt.xlabel("Count")
plt.ylabel("Title")
plt.show()
```
